# Análise Exploratória do Mercado Imobiliário da Cidade de São Paulo

## Contextualização

O mercado imobiliário da cidade de São Paulo é um dos maiores e mais dinâmicos do Brasil. Todos os dias, milhares de imóveis são anunciados para venda, apresentando grande variação de preços em função de fatores como localização, metragem, número de dormitórios, banheiros, vagas de garagem, condomínio e outras características.

Nesse cenário, a análise de dados desempenha um papel importante ao transformar grandes volumes de informações em conhecimento que podem identificar imóveis com melhor custo-benefício, investidores podem encontrar oportunidades de valorização e empresas do setor imobiliário podem compreender melhor o coportamento do mercado.

Esse projeto utiliza uma base de dados pública contendo anúncios de imóveis da cidade de São Paulo com o objetivo de aplicar técnicas de Análise Exploratória de dados (EDA). Durante o desenvolvimento serão utilizadas ferramentas do ecossistema Python para importar, limpar, tratar, explorar e visualizar os dados, buscando identificar padrões, tendências e fatores relacionados ao preço dos imóveis.

Ao final, o projeto apresentará insights que demonstrem como diferentes características influenciam o valor dos imóveis e como essas informações podem contribuir para decisões mais fundamentais.


## Objetivo do Projeto

### Objetivo Central

Realizar uma análise exploratória dos imóveis anunciados na cidade de São Paulo para compreender quais fatores apresentam maior relação com o preço dos imóveis e gerar insights relevantes para o mercado imobiliário.

### Objetivos específicos

    - Importar e compreender a estrutura da base de dados.
    - Identificar e tratar valores ausentes, inconsistências e possíveis duplicidades.
    - REalizar análises estatísticas descritivas.
    - Explorar a distribuição das principais variáveis.
    - Investigar a relação entre características dos imóveis e seus respectivos preços.
    - Produzir visualizações que facilitem a interpretação dos dados.
    - Comunicar os principais resultados de forma clara e objetiva.


## Perguntas de Negócio

Ao longo da análise, busca-se responder às seguintes questões:

    - Qual é o preço médio dos imóveis anunciados?
    - Quais bairros apresentam os maiores e os menores preços médios?
    - Existe relação entre área útil e o preço do imóvel?
    - O número de dormitórios influencia significativamente o valor do imóvel?
    - Quantas vagas de garagem costumam estar associadas aos imóveis mais caros?
    - Existe relação entre o valor do condomínio e o preço do imóvel?
    - Quais características parecem exercer maior influência sobre o valor anunciado?
    - Há indícios de imóveis com preços fora do padrão (outliers)?
    - Existem regi~]oes que oferecem melhor custo por metro quadrado?

## Base de dados

Este projeto utiliza a base de dados House Prices of Sao Paulo City, disponibilizada publicamente na plataforma Kaggle. O conjunto reúne informações sobre imóveis anunciados para venda na cidade de São Paulo, incluindo variáveis como preço, área, número de dormitórios, banheiros, vagas de garagem, condomínio, IPTU e localização.

Link da base de dados: https://www.kaggle.com/datasets/ex0ticone/house-prices-of-sao-paulo-city
