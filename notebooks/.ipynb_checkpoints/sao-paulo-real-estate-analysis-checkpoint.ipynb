# Análise Exploratória do Mercado Imobiliário da Cidade de São Paulo


## 1.0 - Contextualização

O mercado imobiliário da cidade de São Paulo é um dos maiores e mais diversificados do Brasil, reunindo imóveis com diferentes características, localizações e faixas de preço. Essa diversidade torna o processo de avaliação e comparação de imóveis um desafio para compradores, vendedores, investidores e empresas do setor.

Com o crescimento da disponibilidade de dados, técnicas de Análise Exploratória de Dados (Exploratory Data Analysis — EDA) tornaram-se uma ferramenta importante para compreender o comportamento do mercado imobiliário. A partir da análise dos dados, é possível identificar padrões, tendências e relações entre características dos imóveis e seus respectivos valores, fornecendo informações que apoiam a tomada de decisão.

Neste projeto será realizada uma análise exploratória de uma base de dados pública contendo anúncios de imóveis da cidade de São Paulo. O foco será compreender a estrutura dos dados, prepará-los para análise e investigar como fatores como localização, área útil, número de dormitórios, banheiros, vagas de garagem, condomínio e IPTU podem estar relacionados ao preço dos imóveis.

Além de gerar insights sobre o mercado imobiliário, este projeto tem como objetivo demonstrar a aplicação de boas práticas em um fluxo de trabalho de Análise de Dados, incluindo inspeção, preparação, exploração e comunicação dos resultados, utilizando Python e suas principais bibliotecas para ciência de dados.

## 4.0 - Metodologia

Este projeto será desenvolvido seguindo uma adaptação da metodologia CRISP-DM (Cross-Industry Standard Process for Data Mining), amplamente utilizada em projetos de Ciência e Análise de Dados.

O processo está organizado nas seguintes etapas:

    → Compreensão do Negócio (Business Understanding): definição do contexto, dos objetivos e das perguntas de negócio que orientarão toda a análise.
    → Compreensão dos Dados (Data Understanding): carregamento da base de dados, inspeção inicial da estrutura, identificação dos tipos de dados, verificação de valores ausentes, registros duplicados e outras características relevantes.
    → Preparação dos Dados (Data Preparation): tratamento de inconsistências, conversão de tipos de dados, padronização de informações, tratamento de valores ausentes, remoção de duplicidades e demais transformações necessárias para garantir a qualidade da base.
    → Análise Exploratória dos Dados (Exploratory Data Analysis – EDA): aplicação de técnicas estatísticas e visualizações para compreender o comportamento das variáveis, identificar padrões, investigar relações entre os atributos e responder às perguntas de negócio definidas no início do projeto.
    → Conclusão: apresentação dos principais insights obtidos, resposta às perguntas de negócio, discussão das limitações da análise e recomendações para trabalhos futuros.

Todo o desenvolvimento será realizado em ambiente Jupyter Notebook, utilizando Python e bibliotecas especializadas para manipulação, tratamento, análise e visualização de dados, seguindo boas práticas de organização, documentação e reprodutibilidade.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import chardet
```

### 7.2 - Identificando a codificação do arquivo

Antes de realizar a leitura da base de dados, será utilizada a biblioteca **chardet** para identificar automaticamente a codificação (encoding) do arquivo CSV.

### 7.2 - Carregando base de dados

```python
DATA_DIR = Path('../data/raw')
DATA_FILE = DATA_DIR / 'housing_sp_city.csv'
df_raw = pd.read_csv(DATA_FILE, encoding='latin-1')
df = df_raw.copy()
```

### 7.4 - Tipos de dados

Nesta etapa será realizada a inspeção dos tipos de dados de cada variável presente na base.

A identificação correta dos tipos de dados é fundamental para garantir que cada variável seja tratada adequadamente durante a etapa de preparação dos dados. Além disso, essa verificação permite identificar possíveis inconsistências, como datas armazenadas como texto, valores numéricos importados como caracteres ou variáveis categóricas que poderão ser otimizadas posteriormente.

### 7.5 - Verificação de valores ausentes

Valores ausentes são comuns em bases de dados reais e podem ocorrer por diferentes motivos, como falhas na coleta, informações não disponibilizadas ou características que não se aplicam a determinados registros.

Nesta etapa será realizada a identificação da quantidade de valores ausentes em cada variável da base de dados.

O objetivo é compreender a extensão desse problema e subsidiar as decisões que serão tomadas durante a etapa de preparação dos dados. Neste momento, nenhum tratamento será realizado.

Trataremos posteriormente os valores ausentes obedecendo uma regra geral da análise de dados:

    → menos de 5% → geralmente é possível tratar sem grande impacto;
    → entre 5% e 30% → exige uma análise mais cuidadosa;
    → acima de 30% → pode indicar que a variável merece um tratamento específico ou até mesmo ser descartada, dependendo do contexto.

```python
valores_ausentes = (pd.DataFrame({ 'Valores Ausentes': df.isna().sum(), 
                                   'Percentual (%)' : (df.isna().mean() * 100).round(2)})
                    .sort_values(by='Percentual (%)', ascending=False))

valores_ausentes
```

### 7.6 - Verificação de REgistros duplicados

Registros duplicados podem surgir por falhas na coleta, integração de diferentes fontes de dados ou erros durante o processo de armazenamento.

A presença de duplicidades pode comprometer a qualidade das análises, influenciando estatísticas descritivas, distribuições e modelos analíticos.

Nesta etapa será verificada a existência de registros completamente duplicados na base de dados. Caso sejam identificados, sua remoção será avaliada durante a etapa de preparação dos dados.

```python
duplicados = df.duplicated().sum()

print(f'Registros duplicados: {duplicados}')
```

> A inspeção dos registros duplicados indicou que a base possui **1271** registros completamente duplicados. A decisão sobre a remoção desses registros será tomada na etapa de preparação dos dados, considerando o impacto dessa operação na análise.
